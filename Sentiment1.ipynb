```python
import os
import instaloader

L = instaloader.Instaloader()

username = os.getenv("INSTAGRAM_USERNAME")
password = os.getenv("INSTAGRAM_PASSWORD")
if not username or not password:
    raise RuntimeError("Set INSTAGRAM_USERNAME and INSTAGRAM_PASSWORD before running this cell.")

L.login(username, password)

hashtag = instaloader.Hashtag.from_name(L.context, "capetown")
posts = hashtag.get_posts()
for post in posts:
    print(post.date, post.caption)
    break
```
